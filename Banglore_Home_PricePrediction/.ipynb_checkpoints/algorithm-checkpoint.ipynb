```python
import pandas as pd

df1 = pd.read_csv("bengaluru_house_prices.csv")
df1.head()
```

```python
df1.shape
```

<h4 style='color:purple'>1. Drop Unnecessay columns</h4>

```python
df2 = df1.drop(['area_type', 'availability', 'society', 'balcony'], axis="columns")
df2.head()
```

<h4 style='color:purple'>2. Remove rows with Null values</h4>

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3.shape
```

<h4 style='color:purple'>3. Cleaning of size column</h4>
<p>In some row it is given BHK & in some it is show Bedroom so we need to genralize it</p>

```python
df3['bhk'] = df3['size'].apply(lambda x : int(x.split(" ")[0]))
df3.head()
```

<h4 style='color:purple'>4. Cleaning of total_sqft column</h4>
<p>Some rows are given in range format like 200-300, si we need to put average of this </p>

```python
def is_float(x):
    try :
        float(x)
    except :
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)
```

```python
def get_average(x):
    tokens = x.split('-')
    if(len(tokens) == 2):
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(get_average)
df4.head(10)
```

<h4 style='color:purple'>5. Cleaning of Location column</h4>
<p>Here location is categorial parameter. We have around 1300 different location. if we use lable Encoding for each then we will have different 1300 columns which is not reliable. So the solution is we will create common location "Other" which count is less then 10</p>

```python
location_counts = df4['location'].value_counts(ascending=False)
location_counts
```

```python
#location count before categories normalization
len(df4['location'].unique())
```

```python
#get location less then 10 count
location_counts_less_then_10 = location_counts[location_counts<10]

#location count after categories normalization (here we have less row comapre to before)
df4['location'] = df4['location'].apply(lambda x: "other" if x in location_counts_less_then_10 else x)
len(df4['location'].unique())
```

<h4 style='color:purple'>6. Add new column "price_per_sqft"</h4>
<p>for further data cleaning process we need this feature.</p>

```python
df4['price_per_sqft'] = df4['price']*100000/df4['total_sqft']
df4.head()
```

<h4 style='color:purple'>7. Outlier Removal Using Business Logic</h4>
<p>If we look into real life example then we realize size of bedroom should hae minimun threshold(here we have 300 sqft/room). if you find 400 sqft apartment with 2 bhk then we need to remove that row</p>

```python
df4.shape
```

```python
df5 = df4[~(df4['total_sqft']/df4['bhk']<300)]
df5.shape
```

<h3 style='color:blue'>Outlier Removal Using Standard Deviation and Mean</h3>

```python
df5.price_per_sqft.describe()
```

```python
import numpy as np

def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df6 = remove_pps_outliers(df5)
df6.shape
```

<h4 style='color:purple'>8. Data cleaning of bhk columns</h4>
<p>If we check our dataset, we measure that in some case price of 2 bhk is greater then 3 bhk flate in same area. So it will create ambiguty. We need to remove these kind of row</p>

```python
#first we will draw scatter plot to measure price for 2 BHK & 3 BHK plot
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
matplotlib.rcParams["figure.figsize"] = (20,10)

def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
    
plot_scatter_chart(df6,"Rajaji Nagar")
```

```python
plot_scatter_chart(df6,"Hebbal")
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            print(stats['count'])
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return bhk_stats           
    #return df.drop(exclude_indices,axis='index')
remove_bhk_outliers(df6)    
#df7 = remove_bhk_outliers(df6)
# df8 = df7.copy()
#df7.shape
```
